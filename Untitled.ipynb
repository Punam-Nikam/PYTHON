```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
iris=pd.read_csv("C:/Users/punam/OneDrive/Desktop/Book1.csv")
fig,ax=plt.subplots(1,1,figsize(10,8))
iris['species'].value_counts().plot.pie(explode=[0.1,0.1,0.1],autopct='%1.1f%%', ax=ax)
plt.title("Iris Piechart")
plt.show()
```
