```python
def convert_to_sparse(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    
    sparse = []
    sparse.append([rows, cols, 0])  # first row: row, col, non-zero count
    
    count = 0
    for i in range(rows):
        for j in range(cols):
            if matrix[i][j] != 0:
                sparse.append([i, j, matrix[i][j]])
                count += 1
    
    sparse[0][2] = count
    return sparse


# Example matrix
matrix = [
    [0, 0, 3],
    [4, 0, 0],
    [0, 5, 6]
]

sparse = convert_to_sparse(matrix)

print("Sparse Matrix:")
for row in sparse:
    print(row)
```

```python
def merge_arrays(A, B):
    i = j = 0
    merged = []

    while i < len(A) and j < len(B):
        if A[i] < B[j]:
            merged.append(A[i])
            i += 1
        else:
            merged.append(B[j])
            j += 1

    # copy remaining
    merged.extend(A[i:])
    merged.extend(B[j:])
    return merged

A = [1, 4, 7, 9]
B = [2, 3, 8]

print("Merged Array:", merge_arrays(A, B))
```

```python
def matmul(A, B):
    m, p = len(A), len(A[0])
    p2, n = len(B), len(B[0])
    if p != p2: raise ValueError("Incompatible sizes")
    C = [[0]*n for _ in range(m)]
    for i in range(m):
        for j in range(n):
            s = 0
            for k in range(p):
                s += A[i][k] * B[k][j]
            C[i][j] = s
    return C
A=[[1,2],
  [3,4]]
B=[[4,5],
  [6,7]]
matmul(A,B)
```

```python
#Write a program to find the number of vowels, consonants, digits, and white space characters in a string
str1=input("enter a string")

vowels=0
consonants=0
digits=0
space=0
for char in str1:
     if char.lower() in "aeiou":
        vowels+=1
     elif char.isalpha():
        consonants+=1
     elif char.isdigit():
         digits+=0
     elif char.isspace():
         space+=1

print("vowels = ",vowels)
print("consonants =" ,consonants)
print("digits= ",digits)
print("space = ",space)
```
