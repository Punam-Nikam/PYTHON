```python
# Define a list
numbers = [12, 45, 7, 89, 34]

# Initialize the first element as largest
largest = numbers[0]

# Loop through the list
for num in numbers:
    if num > largest:
        largest = num

# Print the largest element
print("The largest element is:", largest)
```

```python
# Define a list

numbers = list(map(int,input("Enter list elements").split()))

# Find the largest element
largest = numbers[0]
# Loop through the list
for num in numbers:
    if num >= largest:
        largest = num
        
# Print the largest element
print("The largest element is:", largest)
```

```python
# Predefined list
numbers = [10, 20, 30, 40, 50]

# Swap first and last elements
numbers[0], numbers[-1] = numbers[-1], numbers[0]

# Print the updated list
print("List after interchanging first and last elements:", numbers)
```

```python
# Initialize empty lists
even_numbers = []
odd_numbers = []

# Loop through numbers from 0 to 150
for num in range(0, 50):
    if num % 2 == 0:
        even_numbers.append(num)
    else:
        odd_numbers.append(num)

# Print the lists
print("Even numbers:", even_numbers)
print("Odd numbers:", odd_numbers)
```

```python
# Take list input from user
numbers = list(map(int, input("Enter list elements separated by space: ").split()))

# Take indices to swap
i = int(input("Enter index of first element to swap: "))
j = int(input("Enter index of second element to swap: "))

# Swap elements using tuple unpacking
# numbers[i], numbers[j] = numbers[j], numbers[i]

temp = numbers[i]
numbers[i] = numbers[j]
numbers[j] = temp

print("List after swapping:", numbers)
```

```python
# Take input for the list
numbers = list(map(int, input("Enter list elements separated by space: ").split()))

# Take input for the element to count
element = int(input("Enter element to count: "))

# Count occurrences using count()
occurrences = numbers.count(element)

print(f"The element {element} occurs {occurrences} times in the list.")
```

```python
# Take input from user
numbers = list(map(int, input("Enter list elements separated by space: ").split()))

# Initialize product
product = 1

# Multiply each element
for num in numbers:
    product *= num

print("The product of all elements in the list is:", product)
```

```python
# Take input from user
string = input("Enter a string: ")

# Reverse the string
reversed_string = string[::-1]

# Check if palindrome
if string == reversed_string:
    print(f"{string} is a palindrome")
else:
    print(f"{string} is not a palindrome")
```

```python

# Take input from user
string = input("Enter a string: ")

# Initialize an empty dictionary
char_count = {}

# Count occurrences of each character
for char in string:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

# Find the character with maximum count
most_frequent = max(char_count, key=char_count.get)
# print(f"The most frequent character is '{most_frequent}' with {char_count[most_frequent]} occurrences.")
print(most_frequent,char_count[most_frequent])
```
