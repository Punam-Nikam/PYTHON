```python
# write a program which swap every all and even position character in a string
```

```python
def swap_odd_even(s):
    s = list(s)   # Convert string to list

    for i in range(0, len(s) - 1, 2):
        s[i], s[i+1] = s[i+1], s[i]   # Swap

    return ''.join(s)   # Convert back to string


# Input from user
string = input("Enter a string: ")
result = swap_odd_even(string)
print("After swapping:", result)
```

```python
def reverse_each_word(s):
    words = s.split()          # built-in method
    result = []

    for word in words:
        result.append(word[::-1])   # slicing (built-in feature)

    return " ".join(result)    # built-in method


# Main program
string = input("Enter a string: ")
print("Output:", reverse_each_word(string))
```

```python
#write a generator function myrange(start,stop,step) which will accept 3 arguments as a start,stop,step and generate a given range .
```

```python
def my_range(start, stop, step):
    if step == 0:
        return

    result = []

    if step > 0:
        while start < stop:
            result = result + [start]
            start = start + step
    else:
        while start > stop:
            result = result + [start]
            start = start + step

    return result

# Example usage
values = my_range(0, 16, 3)
for i in values:
    print(i, end=" ")
```

```python
#write a user defined exception program in python which will find factorial of a number.
# If number is less than zero it should raise the exception as invalid input
```

```python
# user-defined exception
class InvalidInput(Exception):
    pass

def factorial(n):
    if n < 0:
        raise InvalidInput("Invalid Input haii")

    fact = 1
    i = 1
    while i <= n:
        fact = fact * i
        i = i + 1
    return fact


# main program
try:
    num = int(input("Enter a number: "))
    print("Factorial:", factorial(num))
except InvalidInput as e:
    print(e)
```

```python
# b) Write a Python program to create a set by accepting h elements
# (0-9 or A-Z or a-z) input from the user.
# i)
# ii)
# display the set elements and length of the set SI.
# count number of digits, lowercase letters, uppercase letters in a set.
```

```python
# Accept number of elements
n = int(input("Enter number of elements: "))
S1 = set()

# Accept elements from user
for i in range(n):
    ch = input("Enter element (0-9 / A-Z / a-z): ")
    S1.add(ch)

# i) Display set elements and length of the set
print("Set elements:", S1)
print("Length of set:", len(S1))

# ii) Count digits, lowercase and uppercase letters
digits = 0
lower = 0
upper = 0

for ch in S1:
    if '0' <= ch <= '9':
        digits += 1
    elif 'a' <= ch <= 'z':
        lower += 1
    elif 'A' <= ch <= 'Z':
        upper += 1

print("Number of digits:", digits)
print("Number of lowercase letters:", lower)
print("Number of uppercase letters:", upper)
```

```python
# Write  a function to generate Fibonacci sequence of upto first n term
# using recursion.
```

```python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)


def generate_fibonacci(n):
    for i in range(n):
        print(fibonacci(i), end=" ")


# main program
n = int(input("Enter number of terms: "))
generate_fibonacci(n)
```

```python
#  Create a generator which will generate the next prime number from the
# number passed. For example if the number passed is 14, then the next
# Prime number is 17.
```

```python
def next_prime(n):
    num = n + 1

    while True:
        count = 0
        i = 1

        while i <= num:
            if num % i == 0:
                count += 1
            i += 1

        if count == 2:      # prime number condition
            return num

        num += 1


# main program
x = int(input("Enter a number: "))
print("Next prime number is:", next_prime(x))
```

```python
#  What is Package? explain the different ways to import the modules or
# packages.
```

A package in Python is a collection of related modules organized in a directory (folder).
It helps in structuring large programs, improving code reusability, and avoiding name conflicts.

👉 In simple words:

Module = single .py file
Package = folder containing multiple modules

Example structure:

mathpack/
 ├── __init__.py
 ├── add.py
 ├── subtract.py

Different Ways to Import Modules or Packages
1️⃣ Import entire module

Imports the whole module.

import math
print(math.sqrt(16))


✔ Access functions using module_name.function_name

2️⃣ Import module with alias

Used to shorten module name.

import math as m
print(m.sqrt(25))


✔ Improves readability and convenience

3️⃣ Import specific functions or variables

Imports only required members.

from math import sqrt, pow
print(sqrt(9))
print(pow(2, 3))


✔ Saves memory and typing

4️⃣ Import all from a module

Imports everything from a module.

from math import *
print(sqrt(36))


⚠ Not recommended (may cause name conflicts)

5️⃣ Import module from a package

Used when working with packages.

import mathpack.add
print(mathpack.add.sum(5, 3))

6️⃣ Import specific module from a package

Most commonly used package import.

from mathpack import add
print(add.sum(10, 5))

7️⃣ Import specific function from a module inside a package

Direct access to function.

from mathpack.add import sum
print(sum(4, 6))

Conclusion (Exam Line)

Packages help organize multiple related modules, and Python provides multiple ways to import modules or packages depending on usage.

Delegation and Containership with Example
Delegation
Definition:

Delegation is a design concept where one object hands over responsibility to another object to perform a task.

Example:
class Engine:
    def start(self):
        print("Engine started")

class Car:
    def __init__(self):
        self.engine = Engine()

    def start_car(self):
        self.engine.start()

Explanation:

Car does not implement start()

It delegates the task to Engine object

Containership (HAS-A relationship)
Definition:

Containership means one class contains an object of another class.

Example:
class Book:
    def __init__(self, title):
        self.title = title

class Library:
    def __init__(self):
        self.book = Book("Python Programming")

Explanation:

Library has a Book

Book exists as part of Library

Conclusion Line:

Delegation improves code reusability, while containership represents HAS-A relationship between classes.

```python
# d) Multithreaded Program: Square & Cube with Synchronization
# Program:
import threading

lock = threading.Lock()

def square(n):
    with lock:
        print("Square:", n * n)

def cube(n):
    with lock:
        print("Cube:", n * n * n)

t1 = threading.Thread(target=square, args=(5,))
t2 = threading.Thread(target=cube, args=(5,))

t1.start()
t2.start()

t1.join()
t2.join()
```

```python
# a) Python program for BankDemo class

class BankDemo:
    def __init__(self, account_no, name, balance):
        self.account_no = account_no
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print("Amount Deposited:", amount)

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print("Amount Withdrawn:", amount)
        else:
            print("Insufficient Balance")

    def check_balance(self):
        print("Current Balance:", self.balance)


# Main Program
acc = BankDemo(101, "Amit", 5000)
acc.check_balance()
acc.deposit(2000)
acc.check_balance()
acc.withdraw(3000)
acc.check_balance()
```

```python
import threading

lock = threading.Lock()

def ascending():
    with lock:
        for i in range(1, 11):
            print("Ascending:", i)

def descending():
    with lock:
        for i in range(10, 0, -1):
            print("Descending:", i)

t1 = threading.Thread(target=ascending)
t2 = threading.Thread(target=descending)

t1.start()
t2.start()

t1.join()
t2.join()
```

```python
# b) Operator Overloading with Example
# Concept Explanation:

# Operator Overloading allows the same operator to perform different operations depending on operands.
# In Python, this is achieved using magic methods.


class Number:
    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return self.value + other.value


n1 = Number(10)
n2 = Number(20)
print("Addition:", n1 + n2)

# Explanation:

# __add__() overloads + operator

# Enables addition of objects

# Improves code readability
```
