```python
#1.Convert matrix into sparse
def convertSparse(matrix):   ##define function
    rows=len(matrix)
    cols=len(matrix[0])

    sparse=[]   #empty array
    sparse.append([rows,cols,0])    #append no of rows ,no.of cols,and non zero elements AT THE FIRST ROW OF MATRIX

    count=0
    
    for i in range(rows):
        for j in range(cols):
            if matrix[i][j]!=0:
                sparse.append([i,j,matrix[i][j]])
                count+=1

    sparse[0][2]=count
    return sparse


matrix=[
    [0,0,3],
    [4,0,0],
    [0,5,6]
]

sparse=convertSparse(matrix)
print("sparse matrix:")
for row in sparse:
    print(row)
```

```python
# Function to add two sparse matrices
def add_sparse(mat1, mat2):
    result = {}

    # Add elements of first matrix
    for key, value in mat1.items():
        result[key] = value

    # Add elements of second matrix
    for key, value in mat2.items():
        if key in result:
            result[key] += value
        else:
            result[key] = value

    return result


# --- Main Program ---
mat1 = {
    (0, 1): 5,
    (1, 2): 8,
    (2, 0): 3
}

mat2 = {
    (0, 1): 4,
    (1, 0): 7,
    (2, 0): 6
}

print("Sparse Matrix 1:", mat1)
print("Sparse Matrix 2:", mat2)

result = add_sparse(mat1, mat2)

print("\nResultant Sparse Matrix after Addition:")
for key in sorted(result):
    print(f"{key}: {result[key]}")
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    # Insert at beginning
    def insert_begin(self, data):
        new = Node(data)
        new.next = self.head
        self.head = new

    # Insert at end
    def insert_end(self, data):
        new = Node(data)
        if not self.head:
            self.head = new
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new

    # Insert in middle (after a position)
    def insert_middle(self, pos, data):
        new = Node(data)
        temp = self.head
        for _ in range(pos - 1):
            temp = temp.next
        new.next = temp.next
        temp.next = new

    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" → ")
            temp = temp.next
        print("None")


ll = LinkedList()
ll.insert_begin(20)
ll.insert_begin(10)
ll.insert_middle(1, 15)
ll.insert_end(30)
ll.display()
```

```python
# STACK imdernentation using Array with PUSH, POP, traverse, peek operations
stack = []

def push(x):
    stack.append(x)
    print(x, "pushed")

def pop():
    if not stack:
        print("Stack Underflow")
    else:
        print(stack.pop(), "popped")

def peek():
    if not stack:
        print("Stack Empty")
    else:
        print("Top element:", stack[-1])

def traverse():
    if not stack:
        print("Stack Empty")
    else:
        print("Stack elements:")
        for i in reversed(stack):
            print(i)
push(10)
push(20)
push(30)
peek()
pop()
traverse()

```

```python
# 10.	STACK imdernentation using linked list.	
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

top = None  # stack top

def push(x):
    global top
    new = Node(x)
    new.next = top
    top = new
    print(x, "pushed")

def pop():
    global top
    if top is None:
        print("Stack Underflow")
    else:
        print(top.data, "popped")
        top = top.next

def peek():
    if top is None:
        print("Stack Empty")
    else:
        print("Top element:", top.data)

def traverse():
    if top is None:
        print("Stack Empty")
    else:
        print("Stack elements:")
        temp = top
        while temp:
            print(temp.data)
            temp = temp.next
    
push(10)
push(20)
push(30)
peek()
pop()
traverse()
```

```python
# Write an algorithm to reverse the contents of stack using Queue.

stack = [1, 2, 3, 4]  # example stack
queue = []

# Move stack to queue
while stack:
    queue.append(stack.pop())

# Move queue back to stack
while queue:
    stack.append(queue.pop(0))

print("Reversed Stack:", stack)
```

```python
#Write an python program to reverse the nodes from the linked list

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def reverse(head):
    prev = None
    cur = head
    while cur:
        nxt = cur.next
        cur.next = prev
        prev = cur
        cur = nxt
    return prev

def display(head):
    while head:
        print(head.data, end=" ")
        head = head.next
    print()

# create linked list
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = Node(40)

print("Original list:")
display(head)

head = reverse(head)

print("Reversed list:")
display(head)
```
