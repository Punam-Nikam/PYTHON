```python
def convert_sparse(matrix):
    rows=len(matrix)
    cols=len(matrix[0])

    count=0
    sparse=[]
    sparse.append([rows,cols,0])

    for i in range(rows):
        for j in range(cols):
            if matrix[i][j]!=0:
                sparse.append([i,j,matrix[i][j]])
                count+=1

    sparse[0][2]=count
    return sparse

matrix=[
    [0,0,2],
    [3,0,0],
    [0,5,6]
]

sparse=convert_sparse(matrix)
for row in sparse:
    print(row)
```

```python
##ADDITION OF 2 SPARSE MATRICES
def add_sparse(mat1,mat2):
    result= { }
    #add elements of first matrix
    for key,value in mat1.items():
        result[key]=value

    #add elements of second matrix
    for key,value in mat2.items():
        if key in result:
            result[key]+=value
        else:
            result[key]=value
    return result

mat1 = {
    (0, 1) : 3,
    (1, 2) : 4,
    (2, 0) : 2 
    }
mat2 = {
    (0, 1) : 5,
    (1, 2) : 3,
    (2, 0) : 7 
}
print(mat1)
print(mat2)
result=add_sparse(mat1,mat2)
for key in result:
    print(f"{key} : {result[key]}")
```

```python
# STACK implementation using LINKED LISTS

class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

top = None

def push(x):
    global top
    new=Node(x)
    new.next=top
    top=new
    print(x,"pushed")

def pop():
    global top
    if top is None:
        print("Stack underflow")
    else:
        print(top.data,"popped")
        top=top.next

def peek():
    if top is None:
        print("stack is empty")
    else:
        print("Peek :",top.data)

def traverse():
    if top is None:
        print("Stack is empty")
    else:
        print("Stack elements ")
        temp=top
        while temp:
            print(temp.data)
            temp=temp.next

push(10)
push(20)
push(20)
pop()
peek()
traverse()
```

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
    
class linkedlist:
    def __init__(self):
        self.head=None

    def insert_begin(self,data):
        new=Node(data)
        if not self.head:
            self.head=new
            new.next=self.head      # 🔑 last node points to head
            return
        temp=self.head
        while temp.next!=self.head: # 🔑 stop at last node
            temp=temp.next
        new.next=self.head
        temp.next=new               # 🔑 last → new
        self.head=new

    def insert_end(self,data):
        new=Node(data)
        if not self.head:
            self.head=new
            new.next=self.head      # 🔑 circular link
            return
        temp=self.head
        while temp.next!=self.head: # 🔑 condition changed
            temp=temp.next
        temp.next=new
        new.next=self.head          # 🔑 last → head

    def display(self):
        temp=self.head
        if not temp:
            return
        while True:
            print(temp.data,end="=>")
            temp=temp.next
            if temp==self.head:     # 🔑 stop when back to head
                break
        print("(head)")
        
l1=linkedlist()
l1.insert_begin(10)
l1.insert_end(30)
l1.insert_end(20)
l1.display()
```

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
        self.prev=None             # 🔑 extra pointer
    
class dblinkedlist:
    def __init__(self):
        self.head=None

    def insert_begin(self,data):
        newnode=Node(data)
        if self.head:
            self.head.prev=newnode     # 🔑 back link
            newnode.next=self.head
        self.head=newnode

    def insert_end(self,data):
        newnode=Node(data)
        if not self.head:
            self.head=newnode
            return
        temp=self.head
        while temp.next:
            temp=temp.next
        temp.next=newnode
        newnode.prev=temp  # 🔑 back link

    def insert_middel(self,pos,data):
        newnode=Node(data)
        if not self.head:
            return
        temp=self.head
        for _ in range(pos-1):
            temp=temp.next
        newnode.next=temp.next
        temp.next=newnode
        newnode.prev=temp      #🔑🔑
            
    def delete_begin(self):
        if not self.head:
            return
        self.head=self.head.next        # 🔑 move head
        if self.head:
            self.head.prev=None        # 🔑 remove back link

            
    def display(self):
        temp=self.head
        while temp:
            print(temp.data,end="<=>") # 🔑 display style
            temp=temp.next
        print("None")

l1=dblinkedlist()
l1.insert_begin(10)
l1.insert_end(30)
l1.insert_end(20)
l1.delete_begin()
l1.insert_middel(1,15)
l1.insert_middel(3,44)
l1.display()
```
