```python
import threading

lock = threading.Lock()

def ascending():
    with lock:
        for i in range(1, 11):
            print("Ascending:", i)

def descending():
    with lock:
        for i in range(10, 0,-1):
            print("Descending:", i)

t1 = threading.Thread(target=ascending)
t2 = threading.Thread(target=descending)

t1.start()
t2.start()

t1.join()
t2.join()
```
