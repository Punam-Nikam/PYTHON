```python
#4. Write a program to find the number of vowels, consonants, digits, and white space characters in a string. 
s = input("Enter a string: ")
vowel = consonants = digit = whitespace = 0

for ch in s:
    if ch.lower() in 'aeiou': vowel += 1
    elif ch.isalpha(): consonants += 1
    elif ch.isdigit(): digit += 1
    elif ch.isspace(): whitespace += 1

print("Vowels:", vowel)
print("Consonants:", consonants)
print("Digits:", digit)
print("Spaces:", whitespace)
```

```python
#5. Write a program to make a new string with all the consonants 
s = input("Enter a string: ")
new = ""

for ch in s:
    if ch.isalpha() and ch.lower() not in 'aeiou':
        new += ch

print("String with consonants:", new)
```

```python
#10. Python program to find simple interest and compound interest 
principal = float(input("Enter principal amount: "))
rate = float(input("Enter rate of interest: "))
time = float(input("Enter time (in years): "))

simple_interest = (principal * rate * time) / 100
compound_interest = principal * ((1 + rate / 100) ** time - 1)

print("Simple Interest:", simple_interest)
print("Compound Interest:", compound_interest)
```

```python
#11. Python program to check whether a number is Prime or no
n = int(input("Enter a number: "))
if n <= 1:
    print(n, "is not a prime number")
else:
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            print(n, "is not a prime number")
        else:
            print(n, "is a prime number")
```

```python
#12. Python program to Count Even and Odd numbers in a string
# Input string
s = input("Enter a string: ")

Even_cnt = 0
Odd_cnt = 0

for ch in s:
    if ch.isdigit():          # Check if character is a digit
        if int(ch) % 2 == 0:
            Even_cnt += 1
        else:
            Odd_cnt += 1

print("Even digits:", Even_cnt)
print("Odd digits:", Odd_cnt)
```

```python
#18. Write a Python function to reverses a string if its length is a multiple of 5
def reverseof5(s):
    if len(s) % 5 == 0:
        return s[::-1]  # Reverse the string
    else:
        return s

string = input("Enter a string: ")
print("Result:", reverseof5(string))
```
