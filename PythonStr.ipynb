```python
#4. Write a program to find the number of vowels, consonants, digits, and white space characters in a string. 
string=input("Enter string")

vowels=consonants=digits=space=0
for ch in string:

    if ch.lower() in 'aeiou':
        vowels+=1
    elif ch.isalpha():
        consonants+=1
    elif ch.isdigit():
        digits+=1
    elif ch.isspace():
        space+=1
print("vowels",vowels)
print("Consonants",consonants)
print("Digits",digits)
print("Space",space)
```

```python
#12. Python program to Count Even and Odd numbers in a string
cnt=input("enter string")
even=0
odd=0
for i in cnt:
    if i.isdigit():  #check if char is digit
        
        if int(i)%2==0:
            even+=1
        else:
            odd+=1
print("even count=",even)
print("Odd count=",odd)
```
