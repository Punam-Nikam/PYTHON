```python
set1=set()
n=int(input("Enter no of ele in set"))
for i in range(n):
    ele=input("Enter elements of the set")
    set1.add(ele)
print(set1)

print("Length is : ",len(set1))
```

```python
# count no.of digit,lowercase letter,uppercase letter in a set
digit=lcase=ucase=0
for i in set1:
    if i.isdigit():
        digit+=1
    if i.islower():
        lcase+=1
    if i.isupper():
        ucase+=1

print("Digits : ",digit)
print("lowercase : ",lcase)
print("uppercase : ",ucase)
```

```python
def reversestr(string):
    words=string.split()
    s1=[]
    for i in words:
        s1.append(i[::-1])
    return " ".join(s1)
s=input("Enter a string")
print(reversestr(s))
```

```python
def myrange(start,stop,step):
    if step==0:
        return
    result=[]

    if step>0:
        while start<stop:
            result=result+[start]
            start+=step

    else:
        while start>stop:
            result+=[start]
            start+=step

    return result

res=myrange(1,15,2)
print(res)
```

```python
class InvalidInput(Exception):
    pass
    
def factorial(n):
    if n<0:
        raise InvalidInput("Invalid Input ahe")
    else:
        
        fact=1
        i=1
        for i in range(1,n+1):
            fact=fact*i       

    return fact
try:
    n=int(input("Enter a number"))
    print("Factorial : ",factorial(n))
except InvalidInput as e:
    print(e)
```
